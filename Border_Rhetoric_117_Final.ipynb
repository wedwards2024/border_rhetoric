# Introduction: The Politics of Policy: Border Rhetoric on the U.S. House Floor, 2018 - 2022 


In recent years, rhetoric about the border with Mexico has garnered significant attention from U.S. policymakers. Their constituents care about immigration policy. A Pew study ahead of the 2020 election found that 61% of Republican voters and 46% of Democratic voters prioritized immigration when deciding their vote(Pew Research Center, August 2020). The same organization in a report ahead of the 2022 midterms found that 76% of Republican voters and 36% of Democratic voters prioritized this topic (Pew, November 2022). From 2018 - 2022, during the 116th and 117th U.S. Congresses, representatives from both parties spoke on the house floor and framed issues at the border as crises that required severe policy intervention: child separation, drug smugglers, infected migrants. The power balance of Washington tilted between these two periods: during the 116th Congress (2019 - 2021), Republicans held the Presidency and the Senate. Democrats held the House. During the 117th Congress, Democrats held all three. The analysis undertaken in this project shows how border rhetoric on the house floor changed over the course of these two cycles by analyzing transcripts from both periods. 

My initial hypothesis was that my project would show stark changes in border rhetoric on the house floor as party control changed in Washington. I guessed that rhetoric decrying border policies led by the executive branch would prove to be the most salient words in both transcript groups. I expected a clear delineation between the border policy platforms of both main parties, that Democrats in the 116th session criticized Republican policies as cruel and unethical while Republicans in the 117th bemoaned Democratic policies for undermining national security. I predicted that these policy shifts could not explain all the rhetorical shifts that happened – the border and its many crises would be a political cudgel of the party seeking a return to power in Washington.  

My analysis supports my hypothesis that the less powerful party in Washington spoke more about crises at the border, while it does not prove the exact party delineation I expected. During the 116th Congress, the most salient words in the “border” transcripts reflected a case for humane values and welcome attitudes. In the 117th, this perspective was far less common. Instead, the most conspicuous rhetoric argued against the dangers posed by lax border security. One unexpected insight offered by this analysis was the several periods when the border was not mentioned on the house floor - a deeper analysis could explore these gaps. This project offers a first step into a greater analysis of the intersection of policy making and the politics that may slow down policy making.  
 The dataset of urls and associated texts used in this project comes from the digitized transcripts of house proceedings available at the the Library of Congress website. The transcripts vary widely in purpose and length; some are shorter responses to events in the house while others are longer, more purposeful policy advocacy. The analysis implements functions and for-loops while creating visual representations that indicate how and when the word “border” was used in either session. Pandas, WordClouds, and nltk all implemented. The script I used for both Sessions is found in the last section of this project presentation. 


I performed the following methodology and analysis on two datasets, one for each session of Congress. I wrote and ran the python script in a Jupyter Notebook ipynb file. Besides built-in python features and modules like glob, the libraries I used included requests, BeautifulSoup, Pandas, WordCloud, matplotlib, and nltk. The methodology of this analysis consists of two parts: web scraping and text analysis. 
    
The web scraping analysis script is derived from the workshop “Web Scraping Media URLs in Python'' by Rebecca Krisel (workshop). I found the specific set of URLs from a search within the digitized Library of Congress archives of house proceedings available at https://www.congress.gov/. The search specified results from the session of Congress in question. Within that framing, it pulled any transcript that included at least one mention of the word “border” in either the heading or in the body of the transcript. As permitted by the Library, I downloaded these results as a csv file. The csv file has as rows the urls to text files that include the relevant transcripts. The columns of the csv file added the following information to each url in the dataset: the title of the transcript; the date it was said; and the page number and volume in the Congressional record. 

I implemented a function from the Web Scraping workshop to retrieve the transcripts. The function uses requests.get to retrieve the information from the url, encodes it with utf - 8, and returns the html text from the url – the transcript text – as a variable. Once the function was established, I applied it to the data in the cells under the URL column, storing the text data from each url within a single variable, distinguishing each transcript within that variable with an ID index that numbered each url text. I stored this text of every transcript in one file. I then created a for loop to process this text with the html processor BeautifulSoup, clearing away any lingering html code from the text file. The next step was to create a single file for each transcript by identifying each individually according to its number in the larger document. 

The full script I used for each session’s transcript is found at the end of this page. I chose WordClouds and nltk as methods of analysis as they provided efficient means of digestible comparisons between the transcripts of the two periods. The script I used for text analysis was adapted closely from sections of both the Web Scraping workshop and also the Text Analysis Workshop from the CUNY Graduate Center DRI.  (https://gc-dri.github.io/Dhrift-GC/workshops/text-analysis/?page=1). The script for ‘cleaning’ my text files, or making them easier to analyze, depends largely on nltk. The first step was to process out of the text what nltk considers “stopwords,” articles and other frequent words that do not add meaning to the text. My dataframes required additions to the stopwords. Each webpage from the Library contains a phrase at the top explaining the sourcing of the text along with other procedural formalities. For instance, many transcripts include phrases indicating that a member “asked and was given permission to address the House for 1 minute and to revise and extend his remarks” or similar procedural details. Eventually, I was able to identify 71 unimportant words that could clutter this project analysis and added them to the stopwords list. The final steps in cleaning the text, still following the workshop, were to first use nltk to make all words lowercase. Then, to “lemmatize” - a built in nltk feature that makes different words with a shared root word the same, amplifying the presence of a subject that may be represented by varying words in the original text.     

The text analysis involved in this project created three main points of comparisons between the rhetoric of the two sessions: frequency of general border rhetoric over time, language specific to border policies unique to either sessions, and the presence of broader border narratives (advocating sympathy for migrants vs. preventing unauthorized entry). The corresponding scripts are adapted or copied from the mentioned workshops and also Geeks for Geeks (geeksforgeeks.com). To address general rhetoric, I created a function that utilized the count feature in nltk to count the number of appearances of “border” in each text file and add them as a new column in the pandas dataframe. I made a time series with this column as the y-axis and the dates of transcripts as the x-axis. This graph pointed out peaks and troughs in the mentions of “border” throughout the two sessions. These extremes guided my next steps – what caused them? 

To analyze the peaks, I researched attention-grabbing border policies that happened in those time periods. To confirm that these policies were the causes of dense periods of border rhetoric, I implemented a dispersion plot to track appearances of rhetoric about policies, particularly Migrant Protection Protocols, also known as Remain in Mexico, and Title 42. These plots also provided the start of an analysis of my hypothesis that not all rhetoric shifts were easily connected to changing realities at the border. To further investigate this assumption, I utilized the nltk features concordance and similar. Finally, I created word clouds of the processed texts in an attempt to glean a view that least emphasized specific policies.   


The time series indicate the moments during both sessions in which border policies drew representatives’ attention for their significant consequences on the ground - these shifts drew praise and backlash. In the 116th Congress, the largest increase in border rhetoric on the house floor happened in January 2019, when representatives responded to the start of Migration Protection Protocols, also known as Remain in Mexico (Department of Homeland Security, 2019). The border was brought back under the spotlight in July 2019 during negotiations between the Trump Administration and that of Manuel Lopez Obrador to increase Mexican troop presence on the border with Guatemala as a means to prevent migration (NY Post, 2019). As the session continued into 2019, into 2020, and into early 2021, rhetoric about the border tapered off on the house floor. This drop is likely explained by several distractions that happened in 2020: the Black Lives Matter movement, the 2020 election, the January 6th capitol invasion, and most importantly covid-19, which prevented lawmakers from meeting on the house floor for long periods. The first peak in border rhetoric in the 117th sessions happened around April, 2021 - when the Biden began to end Remain in Mexico (NPR, 2021). A year later in April 2022, there was a comparable increase in border conversations on the house floor.  In that time period, the Biden Administration announced its intention to end Title 42, the program that facilitated quick deportations for certain nationals under the guise of sanitary precaution (Texas Tribune, 2022).  

The dispersion plots I used for this project present appearances of words relevant to policies of the session and representatives’ views of the border. In the 116th session, there was consistent rhetoric about families and children at the border. There also appears to be evidence of coordination among representatives projecting a national-security view of the border, as the words “illegal,” “drugs,” “security,” and “wall” appeared together. In the 117th, this security-first border perspective is more consistent throughout. There is mention throughout the transcript of terms relating to the end of Remain in “Mexico” and the end Title “42.” The rhetoric extolling humane values was softer in the 117th session - words like “families” appeared far less often. 

WordClouds also provide evidence for the emergence of a national-security border outlook on the house floor in the 117th session. Both sessions include framing of the humanitarian crisis at the border. A general scan of both WordClouds reveals different understandings of that crisis. In the 116th session, the most salient words that do not appear in the 117th include “provide,” “include,” and “help”, evidence of the sympathetic migrant perspective. In the 117th, terms with a new prevalence include “cartel” and “patrol,” words tied to perceived threats at the border and the need to protect it.     

Features within nltk offer more detail on the politically-fueled shifts in rhetoric between the sessions. The “similar” allows for a comparison between contexts in which “border” appeared in both sessions. “Cartels” and “laws” appear in the 117th, but not in the 116h.

One stark change in border rhetoric between the sessions that is not easily tied to policy is a focus on China. I argue that the 8x increase in “China” mentions is best explained by a political effort to tie border policies to global national security concerns. In the 116th session, there were 5 explicit mentions of China in house border rhetoric. That changed in the 117th.

```python
import requests
def scrape_article(url): #function to web-scrape urls copied from Web Scraping Workshop 
    response = requests.get(url)
    response.encoding = 'utf-8'
    html_string = response.text
    return html_string
```

```python
data_df['text'] = data_df['URL'].apply(scrape_article) #applied scraping function to url column, output in new 'text' column
print(data_df['text']) #confirming scraped html output in text column
```

```python
import nltk
from collections import Counter
cnt = Counter()
from nltk.tokenize import word_tokenize
from nltk.tokenize import word_tokenize
def border_count(text): #a function I made to count the appearances of "border" in each transcript. Adapted from the web scraping function. 
    text_tokens = nltk.word_tokenize(text) 
    nltk_text = nltk.Text(text_tokens)
    count = nltk_text.count('border')
    return count
data_df['count'] = data_df['text'].apply(border_count)#Applied function to text column, added a new "count column"
```

```python
#Now text analysis, adapting Web Scraping Workshop by Rebecca Krisel at https://github.com/rskrisel/web_scraping_workshop
#importing necessary libraries 
import nltk #importing necessary libraries 

#after testing wordclouds, appended stopwords to exclude irrevelant common words shared across transcripts
from nltk.corpus import stopwords 
excess = ['happen','congress','revise','congressional','ask','address','day','speaker','every','permission','publishing','online','today','know','go','tell','remark','give','office','page','biggs','first','record','member','vice','house','sunday','want','well','know','speaker','country','time','unanimous','consent','go','provide','madam','make','united','pro','yield','madam speaker','go','people','make','know','want','say','yield','border', 'house', 'bill','thing','make','think','even','put','use','senate amendment','say','let','one','way','see','u','talk','let one','many','number','say','amendment']
stops = stopwords.words('english')
stops.extend(excess)


from nltk import pos_tag
from nltk.corpus import wordnet 
from nltk.stem.wordnet import WordNetLemmatizer

from wordcloud import WordCloud    
import matplotlib.pyplot as plt

import glob
```

```python
def get_wordnet_pos(word): #implementing a function from workshop to lemmatize text in files
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
from bs4 import BeautifulSoup
with open("all_117_transcript.txt","w") as file: #created a file with partially-clean text of all 117th transcripts
    for text in data_df['text']:
        soup = BeautifulSoup(text)
        transcript = soup.get_text()
        file.write(transcript)
```

```python
! mkdir 117_files_cleaned #made two new folders for cleaned text and wordclouds
! mkdir 117_wordclouds
directory = "117_Congress_transcript"
files = glob.glob(f"{directory}/*.txt")
files #stored pathway to files in 'files' variable
```

```python
id = 0 #adapting a for-loop from the workshop to lemmatize text data 
lexical_density = []  
for filepath in files:
    text = open(filepath, encoding='utf-8').read()
    text_tokens = nltk.word_tokenize(text)
    nltk_text = nltk.Text(text_tokens)
    text_lower = [t.lower() for t in nltk_text if t.isalnum()]
    text_stops = [t for t in text_lower if t in stops]
    text_clean = [WordNetLemmatizer().lemmatize(t, get_wordnet_pos(t)) for t in text_stops]
    id += 1 #storing the cleaned text in a corresponding file within the 117_files_cleaned folder
    with open(f"117_Congress_transcripts/117_transcript_cleaned_{id}.txt", "w") as file:
         file.write(str(text_clean))
```
